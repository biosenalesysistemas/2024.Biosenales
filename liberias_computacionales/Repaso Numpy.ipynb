NumPy (Numerical Python) provides an efficient interface to store and operate on dense data buffers. ​In some ways, NumPy arrays are like Python’s built-in list type, but NumPy arrays provide much more efficient storage and data operations as the arrays grow larger in size.

```python
#import numpy​
import numpy as np
```

We can use np.array to create arrays (ndarrays) from Python lists. NumPy is constrained to arrays that all contain the same type. ​(If types do not match, NumPy will upcast if possible).

```python
np.array([1, 2, 3, 4])
```

For larger arrays, it is more efficient to create arrays from scratch using routines built into NumPy​.

```python
np.zeros(10, dtype='int') # Array with ten entries in zero​
```

```python
np.ones((2, 5), dtype = 'float') #Array, two rows, five columns, all entries with the value one​
```

```python
np.full((2, 5), 3.1416) #two rows, five columns, all entries with the value of 3.1416​
```

**Reshape**

```python
values = np.arange(1, 10) #Array of nine values​
grid  = values.reshape((3, 3)) #Using the method of ndarray
print(values)
print(grid)
print(id(values))
print(id(grid))
```

Note that the size of the initial array must match the size of the reshaped array. Where possible, the reshape method will use a no-copy view of the initial array

**CONCATENATION AND SPLITTING​**

Concatenation, or joining of two arrays in NumPy, is primarily accomplished through the routines np.concatenate, np.vstack, and np.hstack.

```python
x = np.array([1, 2, 3])
y = np.array([3, 2, 1])
np.concatenate([x, y])
```

```python
y
```

Working with arrays of mixed dimensions, it can be clearer to use the np.vstack (vertical stack) and np.hstack (horizontal stack) functions

```python
x = np.array([1, 2, 3])
grid = np.array([[9, 8, 7], [6, 5, 4]])
np.vstack([x, grid]) # vertically stack the arrays​
```

```python
x
```

```python
grid
```

```python
y = np.array([[99], [99]])
np.hstack([grid, y]) # horizontally stack the arrays
```

```python
y
```

The opposite of concatenation is splitting, which is implemented by the functions np.split, np.hsplit, and np.vsplit. ​
For each of these, we can pass a list of indices giving the split points:

```python
x = [1, 2, 3, 99, 99, 3, 2, 1]
x1, x2, x3 = np.split(x, [3, 5])
print(x1, x2, x3)
```

```python
grid = np.arange(16).reshape((4, 4))
upper, lower = np.vsplit(grid, [2])
print(upper)
print(lower)
```

```python
left, right = np.hsplit(grid, [2])
print(left)
print(right)
```

The key to make the computation fast is to use vectorized operations, generally implemented through NumPy’s universal functions (ufuncs).​

```python
np.arange(5) / np.arange(1, 6)
```

```python
np.arange(5)
```

```python
np.arange(1, 6)
```

```python
x = np.arange(9).reshape((3, 3))
print(2 ** x)
```

Computations using vectorization through ufuncs are nearly always more efficient than their counterpart implemented through Python loops, especially as the arrays grow in size.

**COMPARISON OPERATORS**

NumPy also implements comparison operators such as < (less than) and > (greater than) as element-wise ufuncs.​

The result of these comparison operators is always an array with a Boolean data type.

```python
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([5, 4, 3, 2, 1])
```

```python
result_equal = arr1 == arr2
print(result_equal)  # Output: [False False  True False False]
```

```python
result_not_equal = arr1 != arr2
print(result_not_equal)  # Output: [ True  True False  True  True]
```

```python
result_greater_than = arr1 > arr2
print(result_greater_than)  # Output: [False False False  True  True]
```

```python
result_less_than = arr1 < arr2
print(result_less_than)  # Output: [ True  True False False False]
```

```python
np.sum(arr1<6)#To count the number of True entries in a Boolean array
```

```python
np.any(arr2>2)
```

```python
np.all(arr1==2)
```
