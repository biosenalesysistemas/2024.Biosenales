Matplotlib is a multiplatform data visualization library built on NumPy arrays, and designed to work with the broader SciPy stack.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

**Plotting from a scriptâ€‹**

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

Â¡One thing to be aware of!: the plt.show() command should be used only once per Python session, and is most often seen at the very end of the script.

Matplotlib was originally written as a Python alternative for MATLAB users, and much of its syntax reflects that fact.

The MATLAB-style tools are contained in the pyplot (plt) interface.

```python
plt.figure() # create a plot figureâ€‹
# create the first of two panels and set current axisâ€‹

plt.subplot(2, 1, 1) # (rows, columns, panel number)â€‹
plt.plot(x, np.sin(x)) # create the second panel and set current axisâ€‹
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

**OBJECT-ORIENTED INTERFACEâ€‹**

The object-oriented interface is available for these more complicated situations, and for when you want more control over your figure.
Rather than depending on some notion of an â€œactiveâ€ figure or axes, in the object-oriented interface the plotting functions are methods of explicit Figure and Axes objects. â€‹

```python
# First create a grid of plotsâ€‹
# ax will be an array of two Axes objectsâ€‹
fig, ax = plt.subplots(2)
# Call plot() method on the appropriate objectâ€‹
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

If we want to create a single figure with multiple lines, we can simply call the plot function multiple times:â€‹

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

**Adjusting the plot: line colors and stylesâ€‹**



The first adjustment you might wish to make to a plot is to control the line colors and styles. â€‹


To adjust the color, you can use the color keyword, which accepts a string argument representing virtually any imaginable color. â€‹

```python
plt.plot(x, np.sin(x - 0), color='blue') # specify color by nameâ€‹
plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)â€‹
plt.plot(x, np.sin(x - 2), color='0.75') # Grayscale between 0 and 1â€‹
plt.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF)â€‹
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1â€‹
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

If no color is specified, Matplotlib will automatically cycle through a set of default colors for multiple lines.â€‹

You can adjust the line style using the linestyle keyword:

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');
# For short, you can use the following codes:â€‹
plt.plot(x, x + 4, linestyle='-') # solidâ€‹
plt.plot(x, x + 5, linestyle='--') # dashedâ€‹
plt.plot(x, x + 6, linestyle='-.') # dashdotâ€‹
plt.plot(x, x + 7, linestyle=':'); # dotted
```

**Axes Limits**

The most basic way to adjust axis limits is to use the plt.xlim() and plt.ylim() methods:â€‹

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

**LABELING PLOTS**

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend()
```

**BASIC ERRORBARSâ€‹**

A basic errorbar can be created with a single Matplotlib function call:â€‹

```python
plt.style.use('seaborn-whitegrid')
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
plt.errorbar(x, y, yerr=dy, fmt='.k')
```

**HISTOGRAMS, BINNINGS, AND DENSITYâ€‹**

A simple histogram can be a great first step in understanding a dataset.

```python
plt.hist(data, bins=30, alpha=0.5, histtype='stepfilled', color='steelblue', edgecolor='none');
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)
kwargs = dict(histtype='stepfilled', alpha=0.3, normed=True, bins=40)
plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

**MULTIPLE SUBPLOTS**

Sometimes it is helpful to compare different views of data side by side. To this end, Matplotlib has the concept of subplots: groups of smaller axes that can exist together within a single figure.
These subplots might be insets, grids of plots, or other more complicated layouts. In this section, weâ€™ll explore four routines for creating subplots in Matplotlib.

**plt.axes: Subplots by Handâ€‹**

The most basic method of creating an axes is to use the plt.axes function. As weâ€™ve seen previously, by default this creates a standard axes object that fills the entire figure.â€‹

- plt.axes also takes an optional argument that is a list of four numbers in the figure coordinate system. These numbers represent [bottom, left, width, height] in the figure coordinate system, which ranges from 0 at the bottom left of the figure to 1 at the top right of the figure.

```python
#plt.style.use('seaborn-white')
ax1 = plt.axes() # standard axesâ€‹
ax2 = plt.axes([0.65, 0.65, 0.2, 0.2])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4],xticklabels=[], ylim=(-1.2, 1.2))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4],  ylim=(-1.2, 1.2))
x = np.linspace(0, 10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x));
```

```python
for i in range(1, 7):

  plt.subplot(2, 3, i)
  plt.text(0.5, 0.5, str((2, 3, i)),fontsize=18, ha='center')
```

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)

for i in range(1, 7):
  ax = fig.add_subplot(2, 3, i)
  ax.text(0.5, 0.5, str((2, 3, i)), fontsize=18, ha='center')
```
